```python
import pandas as pd

# reding dat
file_path = 'amz_uk_price_prediction_dataset.csv'
df = pd.read_csv(file_path)

# Die ersten Zeilen des Datensets anzeigen
df.head()
```

```python
# Create a frequency table of the categories
category_counts = df['category'].value_counts()

# showing Tables
category_counts
```

```python
#Top 5 Categories
top_5_categories = category_counts.head(5)
top_5_categories
```

```python
import matplotlib.pyplot as plt

# Bar chart for all categories
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar')
plt.title('Verteilung der Produktkategorien')
plt.xlabel('Kategorie')
plt.ylabel('Anzahl der Produkte')
plt.show()
```

```python
# Bar chart for the top 5 categories
plt.figure(figsize=(5, 5))
top_5_categories.plot(kind='bar')
plt.title('Top 5 Produktkategorien')
plt.xlabel('Kategorie')
plt.ylabel('Anzahl der Produkte')
plt.show()
```

```python
# Kreisdiagramm fr die Top 5 Kategorien
plt.figure(figsize=(8, 8))
top_5_categories.plot(kind='pie', autopct='%1.1f%%')
plt.title('Proportion der Top 5 Produktkategorien')
plt.ylabel('')
plt.show()
```

```python
# Calculate mean, median, and mode for the product prices
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]  # mode() returns a series, take the first value

print(f"Mean price: {mean_price}")
print(f"Median price: {median_price}")
print(f"Mode price: {mode_price}")
```

```python
# Calculate variance, standard deviation, range, and interquartile range for the product prices
variance_price = df['price'].var()
std_deviation_price = df['price'].std()
range_price = df['price'].max() - df['price'].min()
iqr_price = df['price'].quantile(0.75) - df['price'].quantile(0.25)

print(f"Variance of price: {variance_price}")
print(f"Standard deviation of price: {std_deviation_price}")
print(f"Range of price: {range_price}")
print(f"Interquartile range of price: {iqr_price}")
```

```python
# Calculate mean, median, and mode for the product ratings
mean_rating = df['stars'].mean()
median_rating = df['stars'].median()
mode_rating = df['stars'].mode()[0]  # mode() returns a series, take the first value

print(f"Mean rating: {mean_rating}")
print(f"Median rating: {median_rating}")
print(f"Mode rating: {mode_rating}")
```

```python
# Calculate variance, standard deviation, and interquartile range for product ratings
variance_rating = df['stars'].var()
std_deviation_rating = df['stars'].std()
iqr_rating = df['stars'].quantile(0.75) - df['stars'].quantile(0.25)

print(f"Variance of rating: {variance_rating}")
print(f"Standard deviation of rating: {std_deviation_rating}")
print(f"Interquartile range of rating: {iqr_rating}")
```

```python
# Calculate skewness and kurtosis for the rating column
skewness_rating = df['stars'].skew()
kurtosis_rating = df['stars'].kurt()

print(f"Skewness of rating: {skewness_rating}")
print(f"Kurtosis of rating: {kurtosis_rating}")
```

```python
# Plot histogram for rating distribution
plt.figure(figsize=(10, 6))
plt.hist(df['stars'], bins=20, edgecolor='black')
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Products')
plt.show()
```
